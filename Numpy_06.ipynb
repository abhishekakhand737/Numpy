## Array Concatenation & Split in NumPy

```python
import numpy as np
```

```python
arr1 = np.arange(1,17).reshape(4,4)
print(arr1)
```

```python
arr2 = np.arange(17,33).reshape(4,4)
print(arr2)
```

```python
arr1 + arr2
```

```python
#It combine arr1 & arr2 element.
np.concatenate((arr1,arr2))
```

```python
list1 = [1,4,3,7]
list2 = [2,5,6,1]
```

```python
list1 + list2
```

```python
#axis represents row                                    
np.concatenate((arr1,arr2), axis = 1)          
```

```python
print(arr1)
print(arr2)
```

```python
#vertically concatenate
np.vstack((arr1,arr2))
```

```python
#horizatly concatenate
np.hstack((arr1,arr2))
```

```python
np.hstack((arr1,arr2,arr1))
```

## Numpy Function :

```python
import numpy as np
```

## arange()

```python
#np.arange(star,end,steps)
arr_1d=np.arange(1,13)
print(arr_1d)
```

```python
even_arr = np.arange(2, 15,2)
print(even_arr)
```

## linspace()

```python
np.linspace(1,6,4)
```

```python
np.linspace(1,158,15)
```

## reshape()

```python
# converting 1d array into 2d 
arr_2d = arr_1d.reshape(3, 4)
print(arr_2d)
```

```python
# converting 1d array into 3d
arr_3d = arr_1d.reshape(2, 3, 2)
print(arr_3d)
```

```python
#import numpy as np
arr = np.arange(1,13).reshape(2,6)
print(arr)
```

## ravel()


```python
# ravel is use to convert 2d arr into 1d
ar.ravel()
```

## flatten()

```python
# same as ravel 
arr.flatten()
```

## transpose()

```python
#converting row into columns
arr.transpose()
```

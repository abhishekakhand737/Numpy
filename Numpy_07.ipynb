## Plot a graph
x_sin = np.arange(0,3*np.pi,0.1)
print(x_sin)

## Graphical Representation of sin function

```python
y_sin = np.sin(x_sin)
print(y_sin)
```

plt.plot(x_sin, y_sin)
plt.show

## Graphical Representation of cos function

```python
y_cos = np.cos(x_sin)
plt.plot(x_sin, y_cos)
plt.show()
```

## Graphical Representation of tan function

```python
y_tan= np.tan(x_sin)
plt.plot(x_sin,y_tan)
plt.show()
```
